# Machine Learning Intro

## High level concepts

### Supervised Learning
- Models that are able to predict labels based on labeled training data

#### Classification
- Supervised learning technique
- Models that are able to predict labels are two of more discrete categories

#### Regression
- Supervised learning technique
- Models that can predict continuous labels

### Unsupervised Learning
- Models that identify structure in unlabeled data

#### Clustering
- Models that are able to detect and identify groups in data

#### Dimensionality Reduction
- Models that detect and identify lower-dimensional structure in higher-dimensional data

## Scikit-Learn estimator API

Steps:

1. Choose a **model class** from scikit-learn
2. Choose model **hyperparameters**
3. **Mutate data** into appropriate format (features matrix and target)
4. **Fit** model to data
5. **Apply** model to new/unknown data (note: for supervised learning predicting new labels is performed with `predict()` method, in unsupervised learning structure is inferred using `predict()` or `transform()` methods

### Unsupervised learning - dimensionality reduction

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components = 2)
```

```python
model.fit(X_iris)
```

```python
X_2D = model.transform(X_iris)
```

```python
X_2D
```

```python
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
seaborn.lmplot("PCA1", "PCA2", hue='species', data=iris, fit_reg=False);
```

### Unsupervised learning - iris clustering

```python
from sklearn.mixture import GMM
```

```python
model = GMM(n_components = 3,
           covariance_type = "full")
```

```python
model.fit(X_iris)
```

```python
y_gmm = model.predict(X_iris)
```

```python
iris['cluster'] = y_gmm
```

```python
iris.head(6)
```

```python
seaborn.lmplot(data = iris,
              hue = "species",
              x = "PCA1",
              y = "PCA2",
              col = "cluster",
              fit_reg = False,
              size = 2.5)
```

## Model Validation Considerations

### A wrong approach

```python
iris.head()
```

```python
X = iris.drop(["PCA1", "PCA2", "cluster", "species"], axis = 1)
```

```python
y = iris["species"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors = 1)
```

```python
model.fit(X, y)
```

```python
y_model = model.predict(X)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y, y_model)
accuracy
```

Model was **trained** and **evaluated** on the **same** data, 100% accuracy should always be approached with caution.

### A better approach, using holdout sets

```python
from sklearn.cross_validation import train_test_split
```

```python
X1, X2, y1, y2 = train_test_split(X, y, random_state = 0, train_size = 0.5)
```

```python
model.fit(X1, y1)
```

```python
y2_model = model.predict(X2)
```

```python
accuracy = accuracy_score(y2, y2_model)
```

```python
accuracy
```

New model has 90% accuracy compared to 100%

### Using cross validation (2-fold)

```python
y1_model = model.fit(X2, y2).predict(X1)
y2_model = model.fit(X1, y1).predict(X2)

accuracy_score(y1, y1_model), accuracy_score(y2, y2_model)
```

This example returns two accuracy scores. It represents 2 fold cross validation, the scores can be averaged out to gather a holistic view of the models 

### Performing 5-fold cross validation using `cross_val_score()`

```python
from sklearn.cross_validation import cross_val_score 
```

```python
cross_val_score(model, X, y, cv = 5)
```
